```python
"""
Project: 
Calculate pivot points for the given stocks.

Background:
Pivot point a method to find support and resistance levels for given stock based on last trading data. 
There are several methods of calculating pivot point this is the simplest one (based on current trading day).

Example:
Attached csv file is contains data of 50 stocks with followind data:

1. Symbol: Name of the stock
2. Open:   Opening price of the given trading day
3. High:   High price for the stock of given trading day
4. Low:    Low price for the stock of given trading day
5. Prev.Close: Previous day closing price of the stock
6. LTP:    Last trading price of the stock of given trading day

Formula:
1. To calculate pivot point following is formula:
   
    Pivot Point (PP) = (High+Low+LTP)/3

2. To calculate first level resistance
    
    Resistance 1 = (2*PP) - Low

3. To calculate second level resistance

    Resistance 2 = PP + (High - Low)

4. To calculate third level resistance

    Resistance 3 = High + 2(PP-Low)

5. To calculate first level support
   
    Support 1 = (2*PP) - High

6. To calculate second level support
   
    Support 2 = PP - (High - Low)

7. To calculate third level support
   
    Support 3 = Low - 2(High - PP)

Assignment:
Import given csv file (with stocks data) and calculate pivot point with above formula for each stocks. 
The Output should be as follows:

(sample)
Symbol    PP        R1       S1       R2        S2       R3       S3
NIFTY50   11770.83  11908.37 11524.32 12154.88  11386.78 12292.42 11140.27
"""
```

```python
# import module
import pandas as pd
```

```python
# create empty dataframe
df = pd.DataFrame()
```

```python
# read csv
df2 = pd.read_csv('NIFTY_50-02022020.csv')
```

```python
# calculate pivot point
df2['PP'] = (df2['High'] + df2['Low'] + df2['LTP']) / 3

# calculate resistance 1
df2['R1'] = (2 * df2['PP']) - df2['Low']

# calculate support 1
df2['S1'] = 2 * df2['PP'] - df2['High']

# calculate resistance 2
df2['R2'] = df2['PP'] + (df2['High'] - df2['Low'])

# calculate support 2
df2['S2'] = df2['PP'] - (df2['High'] - df2['Low'])

# calculate resistance 3
df2['R3'] = df2['High'] + 2 * (df2['PP'] - df2['Low'])

# calculate support 3
df2['S3'] = df2['Low'] - 2 * (df2['High'] - df2['PP'])
```

```python
# add series to empty dataframe
df['Symbol'] = df2['Symbol']
df['PP'] = df2['PP'].round(2)
df['R1'] = df2['R1'].round(2)
df['S1'] = df2['S1'].round(2)
df['R2'] = df2['R2'].round(2)
df['S2'] = df2['S2'].round(2)
df['R3'] = df2['R3'].round(2)
df['S3'] = df2['S3'].round(2)
```
